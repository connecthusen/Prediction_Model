# SGDRegressor

```python
df_feat = df_main.copy()

# Time-based features
df_feat['month']     = df_feat.index.month
df_feat['quarter']   = df_feat.index.quarter
df_feat['year']      = df_feat.index.year
df_feat['trend']     = np.arange(len(df_feat))

# Lag features — 3 lags
df_feat['lag_1']     = df_feat['avg_monthly_price'].shift(1)
df_feat['lag_2']     = df_feat['avg_monthly_price'].shift(2)
df_feat['lag_3']     = df_feat['avg_monthly_price'].shift(3)

# Rolling features — 
df_feat['roll_3']    = df_feat['avg_monthly_price'].shift(1).rolling(3).mean()

# Month sine/cosine encoding
df_feat['month_sin'] = np.sin(2 * np.pi * df_feat['month'] / 12)
df_feat['month_cos'] = np.cos(2 * np.pi * df_feat['month'] / 12)

# Price momentum
df_feat['momentum']  = df_feat['avg_monthly_price'].diff(1)

# Drop NaN rows
df_feat.dropna(inplace=True)

print(f"Rows remaining : {len(df_feat)}")
print(f"Features       : {[c for c in df_feat.columns if c != 'avg_monthly_price']}")
df_feat.head()
```

```python
TARGET = 'avg_monthly_price'
FEATURES = [c for c in df_feat.columns if c != TARGET]

X = df_feat[FEATURES]
y = df_feat[TARGET]

# Last 12 months = test (don't shuffle — time series!)
split = len(df_feat) - 12
X_train, X_test = X.iloc[:split], X.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]

print(f"Train: {X_train.shape} | Test: {X_test.shape}")
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

sgd_model = SGDRegressor(
    loss='squared_error',       # same as Linear Regression loss
    penalty='l2',               # Ridge regularization (prevents overfitting)
    alpha=0.01,                 # regularization strength
    max_iter=1000,              # enough iterations to converge
    learning_rate='adaptive',   # reduces LR when no improvement
    eta0=0.1,                  # initial learning rate
    early_stopping=True,        # stops if val score stops improving
    validation_fraction=0.1,    # 10% of train used for early stop check
    n_iter_no_change=20,        # patience
    random_state=42,
    tol=1e-4
)

sgd_model.fit(X_train_sc, y_train)
print(f"Iterations ran : {model.n_iter_}")

# Feature importance via coefficients
coef_df = pd.DataFrame({
    'Feature'    : FEATURES,
    'Coefficient': sgd_model.coef_
}).sort_values('Coefficient', key=abs, ascending=False)

print("\nFeature Importance :")
print(coef_df.to_string(index=False))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred_train = sgd_model.predict(X_train_sc)
y_pred_test  = sgd_model.predict(X_test_sc)

def evaluate(y_true, y_pred, label):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    r2   = r2_score(y_true, y_pred)
    print(f"\n {label}")
    print(f"  MAE  : {mae:.2f}")
    print(f"  RMSE : {rmse:.2f}")
    print(f"  MAPE : {mape:.2f}%")
    print(f"  R²   : {r2:.4f}")
    return mae, rmse, mape, r2

evaluate(y_train, y_pred_train, "TRAIN")
evaluate(y_test,  y_pred_test,  "TEST")
```

```python
plt.figure(figsize=(14, 5))
plt.plot(y_train.index, y_train, label='Train Actual', color='steelblue')
plt.plot(y_test.index,  y_test,  label='Test Actual',  color='orange')
plt.plot(y_test.index,  y_pred_test, label='Predicted', color='red', linestyle='--', linewidth=2)
plt.title('Linear Regression — Actual vs Predicted', fontsize=14)
plt.legend()
plt.tight_layout()
plt.show()
```

# Prophet

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series, title=''):
    result = adfuller(series.dropna())
    print(f"ADF Test — {title}")
    print(f"  ADF Statistic : {result[0]:.4f}")
    print(f"  p-value       : {result[1]:.4f}")
    print(f"  Critical 1%   : {result[4]['1%']:.4f}")
    print(f"  Critical 5%   : {result[4]['5%']:.4f}")
    if result[1] < 0.05:
        print("  STATIONARY (p < 0.05)")
    else:
        print("   NON-STATIONARY ")

# Test original series
adf_test(df_main['avg_monthly_price'], 'Original Series')
```

```python
prophet_df = pd.DataFrame({
    'ds': df_main.index,
    'y' : df_main['avg_monthly_price']
})

# --- Train Test Split ---
train_prophet = prophet_df.iloc[:int(len(prophet_df)*0.8)]
test_prophet  = prophet_df.iloc[int(len(prophet_df)*0.8):]
```

```python
from prophet import Prophet
# --- Train Model ---
model_P= Prophet(
    seasonality_mode='additive',
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False,
    changepoint_prior_scale=0.01
)
model_P.fit(train_prophet)

# --- Predict ---
future = model_P.make_future_dataframe(
            periods=len(test_prophet), 
            freq='MS')
forecast = model_P.predict(future)
```

```python
predicted = forecast['yhat'].tail(len(test_prophet)).values
actual    = test_prophet['y'].values

print("MAE  :", mean_absolute_error(actual, predicted))
print("RMSE :", np.sqrt(mean_squared_error(actual, predicted)))
print("R2   :", r2_score(actual, predicted))
```

```python
# --- Plot ---
model_P.plot(forecast)
plt.title('Prophet Forecast')
plt.show()
```
